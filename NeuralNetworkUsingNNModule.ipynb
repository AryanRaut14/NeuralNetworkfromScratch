```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AryanRaut14/BreastCancerDetection/main/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
```

```python
df.head()
```

Train Test Split

```python
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

Scaling


```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
y_train
```

Label Encoding

```python
#Since y has output in alphabet, we need to encode it into numeric values.
encoding = LabelEncoder()
y_train = encoding.fit_transform(y_train)
y_test = encoding.transform(y_test)
```

```python
y_train
```

NumPy to PyTorch

```python
x_train_tensor = torch.from_numpy(x_train).float()
x_test_tensor = torch.from_numpy(x_test).float()
y_train_tensor = torch.from_numpy(y_train).float().unsqueeze(1)
y_test_tensor = torch.from_numpy(y_test).float().unsqueeze(1)
```

```python
x_train_tensor.shape
```

```python
y_train_tensor.shape
```

Define Model

```python
import torch.nn as nn

class MySimpleNN(nn.Module):

  def __init__(self, num_features):

    super().__init__()
    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()

  def forward(self, features):
      out = self.linear(features)
      out = self.sigmoid(out)
      return out

  def loss_function(self, y_predicted, y_true):
    return nn.BCELoss()(y_predicted, y_true)
```

Important parameters

```python
learning_rate = 0.1
epochs = 25
```

```python
# define loss function
#binary cross entropy loss
loss_function = nn.BCELoss()
```

Traning pipeline

```python
# create model
n = x_train_tensor.shape[1]
model = MySimpleNN(n).float()

#define loop
for epoch in range(epochs):

  #forward pass
  y_predicted = model(x_train_tensor)

  #calculate loss
  loss = loss_function(y_predicted, y_train_tensor.view(-1,1))

  #backward pass
  loss.backward()

  #update parameters
  with torch.no_grad():
    model.linear.weight -= learning_rate * model.linear.weight.grad
    model.linear.bias -= learning_rate * model.linear.bias.grad

  #zero gradient initialize
  model.linear.weight.grad.zero_()
  model.linear.bias.grad.zero_()

  #print loss in each epoch
  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```

```python
model.linear.weight
```

```python
model.linear.bias
```

Evaluation

```python
with torch.no_grad():
  y_prediction = model.forward(x_test_tensor)
  y_prediction = (y_prediction > 0.95).float()
  accuracy = (y_prediction == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
