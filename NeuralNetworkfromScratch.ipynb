```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/AryanRaut14/BreastCancerDetection/main/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace= True)
```

```python
df.head()
```

Train Test Split

```python
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

Scaling


```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
y_train
```

Label Encoding

```python
#Since y has output in alphabet, we need to encode it into numeric values.
encoding = LabelEncoder()
y_train = encoding.fit_transform(y_train)
y_test = encoding.transform(y_test)
```

```python
y_train
```

NumPy to PyTorch

```python
x_train_tensor = torch.from_numpy(x_train)
x_test_tensor = torch.from_numpy(x_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
x_train_tensor.shape
```

```python
y_train_tensor.shape
```

Define Model

```python
class MySimpleNN():

  def __init__(self, x):

    self.weights = torch.rand(x.shape[1], 1, dtype=torch.float64, requires_grad=True)
    self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)

  def forward(self, x):
      z = torch.matmul(x , self.weights) + self.bias
      y_predicted = torch.sigmoid(z)
      return y_predicted

    #one of the possible approaches
    #def loss(self, y_predicted, y):
    #  loss = torch.mean((y - y_predicted))**2

  def loss_function(self, y_predicted, y):
      #clamp predictions to avoid log(0)
      epsilon = 1e-7
      y_predicted = torch.clamp(y_predicted, epsilon, 1-epsilon)

      #calculate loss
      loss = -(y_train_tensor * torch.log(y_predicted) + (1 - y_train_tensor) * torch.log(1 - y_predicted)).mean()
      return loss
```

Important parameters

```python
learning_rate = 0.1
epochs = 25
```

Traning pipeline

```python
# create model
model = MySimpleNN(x_train_tensor)

#define loop
for epoch in range(epochs):

  #forward pass
  y_predicted = model.forward(x_train_tensor)

  #calculate loss
  loss = model.loss_function(y_predicted, y_train_tensor)

  #backward pass
  loss.backward()

  #update parameters
  with torch.no_grad():
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

  #zero gradient initialize
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  #print loss in each epoch
  print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```

Evaluation

```python
with torch.no_grad():
  y_prediction = model.forward(x_test_tensor)
  y_prediction = (y_prediction > 0.95).float()
  accuracy = (y_prediction == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```
